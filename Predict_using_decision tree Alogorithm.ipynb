```python
import sklearn.datasets as datasets
import pandas as pd
```

Loading the dataset and print head of the dataset

```python
dataSet=datasets.load_iris()
df=pd.DataFrame(dataSet.data,columns=dataSet.feature_names)
print(df.head(5))
```

# Check shape of the data

```python
df.shape
```

# Check null values of the data

```python
df.isna().sum()
```

# Check the information of the data

```python
df.info()
```

# Overall description of the dataset

```python
df.describe()
```

# Correlation between columns


```python
df.corr()
```

# fearures and labels

```python
x=df
y=dataSet.target
print(y)
```

# Split the dataset in training and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

# Train the model on train dataset

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
print("Model trained succesfully!!!")
```

# Define prediction variable y_prdiction

```python
y_predict=dtree.predict(x_test)
y_predict
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```

# The overall report of Matrix

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_predict,digits=3))
```

# Visual Representation

```python
from sklearn import tree
import matplotlib.pyplot as plt
fn=["sepal length(cm)","sepal width(cm)","petal length(cm)","sepal width(cm)"]
cn=["setosa","versicolor","viginica"]
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(dtree,feature_names=fn,class_names=cn,filled=True)
```
